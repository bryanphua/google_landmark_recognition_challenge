### after downloading the images on train.csv we check the number of successful downloads

```python
log_path = "/mount_point/home/ubuntu/.kaggle/competitions/landmark-recognition-challenge/data/download_train_log.csv"
with open(log_path,'r') as log:
    entry = log.readline()
    success = 0
    fail = 0
    count = 0
    while entry:
        entry = entry[:-1].split(',')
        if len(entry)>3:
            if entry[3] == 'Y':
                success += 1
            if entry[3] == 'N':
                fail += 1

        count += 1
        entry = log.readline()
```

```python
print("TOTAL DOWNLOADS: {}".format(success+fail))
print("SUCCESSFUL DOWNLOADS: {}".format(success))
print("FAILED DOWNLOADS: {}".format(fail))
print("FAILURE RATE: {}".format(float(fail)/(success+fail)))
```

## Analyzing training data

```python
train_log_path = "/mount_point/home/ubuntu/.kaggle/competitions/landmark-recognition-challenge/data/download_train_log.csv"
train_df = pd.read_csv(train_log_path,header=1,skiprows=0)
train_df.head()
```

```python
# filter out failed downloads
train_df = train_df.loc[train_df['download']=='Y']
```

### Image distribution among classes/landmarks

```python
label_frequency = train_df['landmark_id'].value_counts()
label_frequency[0:50].plot(kind='bar')
plt.suptitle('top 50 landmarks'),plt.xlabel('landmark_id'),plt.ylabel('frequency')
plt.show()
```

```python
label_frequency[-50:].plot(kind='bar')
plt.suptitle('bottom 50 landmarks'),plt.xlabel('landmark_id'),plt.ylabel('frequency')
plt.show()
```

```python
label_frequency.plot(kind='box',sym='')
plt.ylabel('frequency')
plt.show()
print(label_frequency.describe())
```

large skew of images towards top 25 percentile

## Images for top 10 landmarks

```python
from IPython.display import Image, display
num_images = 5
num_classes = 10

fig = plt.figure(figsize=(15,35))
plt.subplots_adjust(wspace=0.01, hspace=0.01)
for i in range(num_classes):
    landmark_id = label_frequency.index[i]
    lm_df = train_df.loc[train_df['landmark_id'] == landmark_id]
    interval = int(lm_df.shape[0]/(num_images+1))

    for j in range(num_images):
        path = lm_df.iloc[interval*j]['path']
        img = mpimg.imread(path)
        ax = plt.subplot(num_classes,num_images,num_images*i+j+1)
        ax.set_xticks([])
        ax.set_yticks([])
        ax.set_title(landmark_id)
        plt.imshow(img)
        
plt.show()
```

Images may be black and white, rotated, different lighting conditions, different viewing angle including from INSIDE the landmark building

## and bottom 10 landmarks

```python
num_rows = 2
num_cols = 5
num_classes = 10

fig = plt.figure(figsize=(15,7))
plt.subplots_adjust(wspace=0.01, hspace=0.01)
for i in range(num_rows):
    for j in range(num_cols):
        landmark_id = label_frequency.index[-1-i*num_cols-j]
        lm_df = train_df.loc[train_df['landmark_id'] == landmark_id]
        path = lm_df.iloc[0]['path']
        img = mpimg.imread(path)
        ax = plt.subplot(num_rows,num_cols,num_cols*i+j+1)
        ax.set_xticks([])
        ax.set_yticks([])
        ax.set_title(landmark_id)
        plt.imshow(img)
        
plt.show()
```
